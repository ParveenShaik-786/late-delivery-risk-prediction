```python
### Data Leakage Analysis

# Potential Leakage Columns Identified:

# 1. Delivery Status
# 2. Days for shipping (real)

# Reason:

# These columns contain information that becomes available only after shipment completion and are strongly related to the target variable.

# Using these features during model training would cause data leakage and produce unrealistically high model performance.

# These columns should be excluded from machine learning modeling.
```

```python
# Numerical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:", len(num_cols))
print("Categorical Columns:", len(cat_cols))
```
